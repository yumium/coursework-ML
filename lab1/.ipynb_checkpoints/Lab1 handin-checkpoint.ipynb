```python
import _pickle as cp
import numpy as np
import matplotlib.pyplot as plt

X, y = cp.load(open('winequality-white.pickle', 'rb'))

N, D = X.shape
X = np.column_stack((np.ones(N),X)) # add bias term
D += 1

N_train = int(0.8 * N)
N_test = N - N_train
X_train = X[:N_train]
y_train = y[:N_train]
X_test = X[N_train:]
y_test = y[N_train:]
```

### Handin 1
Distribution of the qualities of wine.
x-axis: Quality level, y-axis: Frequency

```python
import collections

def plotBar(npArr):
    count = collections.Counter(npArr)
    xs = []
    ys = []
    for k in sorted(count.keys()):
        xs.append(k)
        ys.append(count[k])
    plt.bar(xs, ys)
    plt.show()
```

```python
plotBar(y)
```

### Handin 2
Average quality and mean squared error using this predictor

```python
avgPredict = sum(y_train) / y_train.size
print("The average quality is {}".format(avgPredict))

meanSqrErrorAvgPredictTrain = sum([(q-avgPredict)**2 for q in y_train]) / y_train.size
meanSqrErrorAvgPredictTest = sum([(q-avgPredict)**2 for q in y_test]) / y_test.size
print("The mean squared error on training data using the avg. of y is {}".format(meanSqrErrorAvgPredictTrain))
print("The mean squared error on test data using the avg. of y is {}".format(meanSqrErrorAvgPredictTest))
```

### Handin 3
We standardize the input data and train the model using the closed-form solution of the LSE.

It's not necessary to standardize here because multiplier to the data would not affect the result here (unlike in Ridge  Regression). w will adjust so that it gives the same prediction as the 
non-standardized version

```python
def mean(npArr):
    return sum(npArr) / npArr.size

def var(npArr):
    xBar = mean(npArr)
    return sum([(x-xBar)**2 for x in npArr]) / npArr.size

for i in range(1,D):    # skip bias term
    mu = mean(X_train[:,i])
    sigma2 = var(X_train[:,i])
    X_train[:,i] = (X_train[:,i]-mu) / (sigma2**0.5)    # Apply mu and sigma of training data to standardize all data
    X_test[:,i] = (X_test[:,i]-mu) / (sigma2**0.5)
```

```python
w = np.linalg.inv(X_train.transpose() @ X_train) @ X_train.transpose() @ y_train
y_pred_train = X_train @ w
y_pred_test = X_test @ w

meanSqrErrLinRegTrain = sum([(y_pred_train[i] - y_train[i])**2 for i in range(y_train.size)]) / y_train.size
meanSqrErrLinRegTest = sum([(y_pred_test[i] - y_test[i])**2 for i in range(y_test.size)]) / y_test.size

print("The mean squared error on training data using least squares estimate is {}".format(meanSqrErrLinRegTrain))
print("The mean squared error on test data using least squares estimate is {}".format(meanSqrErrLinRegTest))
```

### Handin 4
Here the change in training and test data with respect to increase in training data

The red line is test error, the blue line is the training error

The model is just right, as the test error has decreased to a point where the gradient is near horizontal.

More data is not necessary to get optimal test error. From the graph, we can see that 300 data points is more than enough to reduce test error to its optimal level.

```python
training_error = []
test_error = []
n_data_points = []

for i in range(20,600,20):
    X_seg = X_train[:i,:]
    w = np.linalg.inv(X_seg.transpose() @ X_seg) @ X_seg.transpose() @ y_train[:i]
    y_pred_train = X_seg @ w
    y_pred_test = X_test @ w
    training_error.append(sum([(y_pred_train[i] - y_train[i])**2 for i in range(y_pred_train.size)]) / y_pred_train.size)
    test_error.append(sum([(y_pred_test[i] - y_test[i])**2 for i in range(y_test.size)]) / y_test.size)
    n_data_points.append(i)

plt.plot(n_data_points,training_error,"b-",n_data_points,test_error,"r-")
```
