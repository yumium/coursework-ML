```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0,2*np.pi,100)
f = np.sin(x)
plt.plot(f)
plt.show()
```

```python
A = np.linspace(0,3,10)
plt.plot(A,A,"b-",A,A**2,"g^",A,A**3,"ro")
plt.show()
```

```python
names = ['group_a', 'group_b', 'group_c']
values = [1, 10, 100]

plt.figure(figsize=(12, 5))

plt.subplot(131)
plt.bar(names, values)
plt.xlabel("groups")
plt.ylabel("frequency")
plt.subplot(132)
plt.scatter(names, values)
plt.grid(True)
plt.subplot(133)
plt.plot(names, values)
plt.suptitle(r'$\mu = 5$')
plt.show()
```
